```python
from pylabel import importer
```

```python
path_to_annotations = r"C:\Users\Sathwik\Downloads\ev\annotations\bbox-annotations.json"
path_to_images = r"C:\Users\Sathwik\Downloads\ev\images"
```

```python
dataset = importer.ImportCoco(path_to_annotations, path_to_images=path_to_images, name="bbox_coco")
```

```python
dataset.df['cat_id'] = dataset.df['cat_id'].map({'1' : '0', '2' : '1'})
```

```python
print(f"Number of images: {dataset.analyze.num_images}")
print(f"Number of classes: {dataset.analyze.num_classes}")
print(f"Classes:{dataset.analyze.classes}")
print(f"Class counts:\n{dataset.analyze.class_counts}")
print(f"Path to annotations:\n{dataset.path_to_annotations}")
```
